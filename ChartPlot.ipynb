```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import joypy

# Cargar el dataset localmente
co2_data = pd.read_csv("annual-co2-including-land-use.csv")

co2_data
```

```python
unique_countries = co2_data["Entity"].unique()
unique_countries
```

```python
# Seleccionar paises para visualizar
selected_countries = ["United States", "China", "India", "European Union (27)", "Asia (excl. China and India)"]
co2_selected = co2_data[co2_data["Entity"].isin(selected_countries)]

# Crear un barplot agrupado por país y año
plt.figure(figsize=(14, 7))
sns.barplot(data=co2_selected, x="Year", y="Annual CO₂ emissions including land-use change", hue="Entity", ci=None, palette="rocket")
plt.title("Emisiones de CO2 por Países a lo largo del Tiempo", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=14)
plt.legend(title="País")
plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_total_years.png"
plt.savefig(ruta)

plt.show()
```

```python
# Filtrar por algunos países específicos
selected_countries = ["United States", "China", "India", "European Union (27)", "Asia (excl. China and India)"]


# Filtrar para el año 2020 y seleccionar algunos países
co2_bar_data = co2_data[(co2_data["Year"] == 2020) & (co2_data["Entity"].isin(selected_countries))]

# Crear un barplot con estilos mejorados
plt.figure(figsize=(12, 6))
sns.barplot(data=co2_bar_data, x="Entity", y="Annual CO₂ emissions including land-use change", palette="rocket", ci=None)
plt.title("Emisiones de CO2 por Países en 2020", fontsize=16)
plt.xlabel("País", fontsize=14)
plt.ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_2020.png"
plt.savefig(ruta)

plt.show()
```

```python
# Filtrar para obtener solo los últimos 10 años
latest_year = co2_data["Year"].max()
co2_recent = co2_data[co2_data["Year"] >= (latest_year)]

# Filtrar por países seleccionados
co2_selected = co2_recent[co2_recent["Entity"].isin(selected_countries)]

# Crear un gráfico de barras apiladas por año
plt.figure(figsize=(12, 7))
sns.histplot(
    data=co2_selected,
    x="Year",
    weights="Annual CO₂ emissions including land-use change",
    hue="Entity",
    multiple="stack",
    palette="rocket")
plt.title("Emisiones de CO2 por Países en los Últimos 10 Años", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=14)
plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_2020_1.png"
plt.savefig(ruta)

plt.show()
```

```python
# Filtrar para los últimos 10 años
latest_year = co2_selected["Year"].max()
co2_recent = co2_selected[co2_selected["Year"] >= (latest_year - 9)]

# Crear un barplot agrupado por país y año
plt.figure(figsize=(14, 7))
sns.barplot(
    data=co2_recent,
    x="Year",
    y="Annual CO₂ emissions including land-use change",
    hue="Entity",
    ci=None,
    palette="rocket")
plt.title("Emisiones de CO2 por Países a lo largo del Tiempo (Últimos 10 Años)", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=14)
plt.legend(title="País")
plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_10años_5.png"
plt.savefig(ruta)

plt.show()
```

```python
# Filtrar para obtener solo los últimos 10 años
latest_year = co2_data["Year"].max()
co2_recent = co2_data[co2_data["Year"] >= (latest_year-9)]


# Filtrar por países seleccionados
co2_selected = co2_recent[co2_recent["Entity"].isin(selected_countries)]

# Crear un gráfico de barras apiladas por año
plt.figure(figsize=(12, 7))
sns.histplot(
    data=co2_selected,
    x="Year",
    weights="Annual CO₂ emissions including land-use change",
    hue="Entity",
    multiple="stack",
    palette="rocket")
plt.title("Emisiones de CO2 por Países en los Últimos 10 Años", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=14)
plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_10años_1.png"
plt.savefig(ruta)


plt.show()
```

```python
# Filtrar para los últimos 10 años
latest_year = co2_selected["Year"].max()
co2_recent = co2_selected[co2_selected["Year"] >= (latest_year - 150)]


# Obtener la paleta de colores "rocket"
palette = sns.color_palette("rocket", n_colors=len(selected_countries))

# Crear un gráfico de barras para cada país
plt.figure(figsize=(12, 12))
for idx, country in enumerate(selected_countries):
    plt.subplot(len(selected_countries), 1, idx + 1)
    country_data = co2_recent[co2_recent["Entity"] == country]
    
    # Dibujar las barras
    plt.bar(country_data["Year"], country_data["Annual CO₂ emissions including land-use change"], color=palette[idx], alpha=0.8, width=0.8, label=country)
    plt.fill_between(
        country_data["Year"], 0, country_data["Annual CO₂ emissions including land-use change"], color=palette[idx], alpha=0.3)
    
    plt.title(f"Emisiones de CO2 para {country}", fontsize=14)
    plt.xlabel("Año")
    plt.ylabel("Emisiones de CO2 (millones de toneladas)")
    plt.legend()

plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_10años_split.png"
plt.savefig(ruta)

plt.show()
```

```python
# Definir una paleta de colores
color_palette = sns.color_palette("rocket", len(selected_countries))

# Crear una figura todos los gráficos
fig, axes = plt.subplots(nrows=len(selected_countries), ncols=1, figsize=(10, 12))

# Iterar sobre cada país
for ax, country, color in zip(axes, selected_countries, color_palette):
    country_data = co2_recent[co2_recent["Entity"] == country]
    sns.lineplot(
        data=country_data,
        x="Year",
        y="Annual CO₂ emissions including land-use change",
        ax=ax,
        marker='o',
        color=color
    )
    ax.set_title(f"Emisiones de CO2 en {country} a lo largo del Tiempo", fontsize=14)
    ax.set_xlabel("Año", fontsize=12)
    ax.set_ylabel("Emisiones de CO2 (millones de toneladas)", fontsize=12)

plt.tight_layout()

# Guardar la imagen en carpeta local
ruta = "imagenes/barplot_CO2_10años_lineas.png"
plt.savefig(ruta)

plt.show()
```
